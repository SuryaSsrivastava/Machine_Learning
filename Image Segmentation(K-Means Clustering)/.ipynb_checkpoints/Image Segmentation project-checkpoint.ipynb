```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image=cv2.imread("./close-up-image-of-the-beautiful-vibrant-orange-royalty-free-image-1580855109.jpg")
```

```python
image.shape
```

```python
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
orig_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(orig_image)
plt.axis("off")
plt.show()
```

```python
orig_image.shape
```

```python
# Flatten image
fl_img=orig_image.reshape((-1,3))
print(fl_img.shape)
```

### Extracting Dominant Colors

```python
from sklearn.cluster import KMeans
```

```python
dominant_color=4
km=KMeans(n_clusters=dominant_color)
km.fit(fl_img)
```

```python
centers=km.cluster_centers_
print(centers)
```

```python
centers=np.array(centers,dtype='uint8')
print(centers)
```

```python
labels=km.labels_
print(labels)
```

```python
fig,ax=plt.subplots(1,4,figsize=(8,4))
colors=[]
for i,each_col in enumerate(centers):
    colors.append(each_col)
    a=np.zeros((100,100,3),dtype='uint8')
    a[:,:,:]=each_col
    ax[i].imshow(a)
    ax[i].axis("off")
```

```python
for ix in range(fl_img.shape[0]):
    fl_img[ix]=colors[labels[ix]]
fl_img=fl_img.reshape((orig_image.shape))
plt.imshow(fl_img)
plt.axis("off")
plt.show()
```
