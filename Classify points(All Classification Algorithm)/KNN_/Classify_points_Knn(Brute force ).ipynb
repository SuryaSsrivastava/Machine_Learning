```python
### 1. Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx=pd.read_csv("../Logistic_X_Train.csv")
dfy=pd.read_csv("../Logistic_Y_Train.csv")
df_test=pd.read_csv("../Logistic_X_Test.csv")
```

```python
dfx.head()
```

```python
dfy.head(2)
```

```python
X_train=dfx.values
Y_train=dfy.values
X_test=df_test.values
Y_train=Y_train.reshape(-1,)
```

```python
print(X_train.shape,Y_train.shape)
print(X_test.shape)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=Y_train);
```

```python
def distance(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))


def KNN(X_train,Y_train,query_point,k=20):
    
    dist=[]
    
    m=X_train.shape[0]
    
    x2=query_point
    for i in range(m):
        x1=X_train[i]
        d=distance(x1,x2)
        y=Y_train[i]
        dist.append((d,y))
        
    dist.sort()
    dist=dist[:k+1]
    dist=np.array(dist)
    new_=np.unique(dist[:,1],return_counts=True)
    predict=np.argmax(new_[1])
    return predict
```

```python
predict=[]
for i in range(X_test.shape[0]):
    predict.append(KNN(X_train,Y_train,X_test[i]))
```

```python
knn_=pd.DataFrame({"label":predict})
knn_.to_csv("KNN_scratch.csv",index=False)
```

### Using gridSearch on KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
knn_c=KNeighborsClassifier()
print(knn_c)
```

```python
params={
    "n_neighbors":[15,20,25,30,35,40],
    "algorithm":['ball_tree', 'kd_tree', 'brute']
}
```

```python
gd=GridSearchCV(estimator=knn_c,param_grid=params,n_jobs=-1)
gd.fit(X_train,Y_train)
```

```python
knn_grid=gd.predict(X_test)
```
