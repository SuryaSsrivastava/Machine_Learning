```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
dfx=pd.read_csv("../Logistic_X_Train.csv")
dfy=pd.read_csv("../Logistic_Y_Train.csv")
df_test=pd.read_csv("../Logistic_X_Test.csv")
```

```python
dfx.head()
```

```python
X_train=dfx.values
Y_train=dfy.values
X_test=df_test.values
Y_train=Y_train.reshape(-1,)
```

```python
print(X_train.shape,Y_train.shape)
print(X_test.shape)
```

```python
def custom_kernal(x1,x2):
    return np.square(np.dot(x1,x2.T))
```

```python
x3=custom_kernal(X_train[:,0],X_train[:,1])
X_=np.zeros((X_train.shape[0],3))
X_[:,0]=X_train[:,0]
X_[:,1]=X_train[:,1]
X_[:,2]=x3
print(X_.shape)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=Y_train);
```

```python
def plot3D(x,color,show=False):
    fig,ax=plt.subplots(1,figsize=(7,7))
    ax=plt.axes(projection='3d')
    x1=x[:,0]
    x2=x[:,1]
    x3=x[:,2]
    ax.scatter(x1,x2,x3,zdir='z',c=color,s=20,depthshade=True)
    if show==True:
        plt.show()
    return ax
    
```

```python
plot3D(X_,Y_train,True)
```
